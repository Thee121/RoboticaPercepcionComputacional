```python
import cv2
import numpy as np
import select_pixels as sel
from imageio import imsave
from matplotlib import pyplot as plt
from pathlib import Path
base_dir = Path.cwd()
media_path = base_dir.parent / 'Media'
```

```python
capture = cv2.VideoCapture(media_path / 'videos/videoB.avi')

key = 0
while (key != 113):
    # Leo una imagen del video
    ret,im = capture.read()
 
    if not ret:
        print('Se acabo el video.')
        cv2.destroyAllWindows()
        break
    
    # Save last valid frame
    last_valid_frame = im.copy()  
      
    # Muestro la imagen en una ventana OpenCV
    cv2.imshow("Captura", im)

    key = cv2.waitKey(35)
    
# Cierro la ventana 
cv2.destroyAllWindows() 
capture.release()
```

```python
# Transformo la imagen al formato Python
imNumpy=cv2.cvtColor(last_valid_frame, cv2.COLOR_BGR2RGB)

# Llamo a la funcion para etiquetar la imagen
markImg = sel.select_fg_bg(imNumpy)

plt.imshow(markImg)
plt.show()
```

```python
# Guardo las imagenes
imsave('imgb.png',imNumpy)
imsave('imgb_mark.png',markImg)
```

```python
#Captura de imagenes
capture = cv2.VideoCapture(media_path / 'videos/videoB.avi')
key = 0

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('videoB_normalizado.avi', fourcc, 25.0, (320, 240))

# We do quit if the keyword "q" = 113 is intered.
while (key != 113):
    # Leo una imagen del video
    ret,im = capture.read()
  
    if not ret:
        print('Se acabo el video.')
        cv2.destroyAllWindows()
        break
        
    # Normalize image
    normalized_image = cv2.normalize(im, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
    
    # Guardar la imagen normalizada en el video
    out.write(np.uint8(normalized_image))
    
    # Mostrar la imagen normalizada en una ventana
    cv2.imshow("Captura", normalized_image)

    key = cv2.waitKey(35)
    
# Cierro la ventana 
cv2.destroyAllWindows() 
capture.release()
out.release()
```
